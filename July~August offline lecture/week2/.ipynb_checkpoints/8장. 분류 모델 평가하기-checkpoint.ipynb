```python
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import average_precision_score
from matplotlib import pyplot
# generate 2 class dataset
X, y = make_classification(n_samples=1000, n_classes=2, weights=[1,1], random_state=1)
# split into train/test sets
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.5, random_state=2)
# fit a model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(trainX, trainy)
# predict probabilities
probs = model.predict_proba(testX)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# predict class values
yhat = model.predict(testX)
# calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(testy, probs)
# calculate F1 score
f1 = f1_score(testy, yhat)
# calculate precision-recall AUC
auc = auc(recall, precision)
# calculate average precision score
ap = average_precision_score(testy, probs)
print('f1=%.3f auc=%.3f ap=%.3f' % (f1, auc, ap))
# plot no skill
pyplot.plot([0, 1], [0.5, 0.5], linestyle='--')
# plot the precision-recall curve for the model
pyplot.plot(recall, precision, marker='.')
# show the plot
pyplot.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot

# generate 2 class dataset
X, y = make_classification(n_samples=1000, n_classes=2, weights=[1,1], random_state=1)
# split into train/test sets
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.5, random_state=2)
# fit a model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(trainX, trainy)
# predict probabilities
probs = model.predict_proba(testX)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(testy, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(testy, probs)
# plot no skill
pyplot.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
pyplot.plot(fpr, tpr, marker='.')
# show the plot
pyplot.show()

# pyplot.savefig('roc.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels


def main():
    # Wine data set  X, y  .
    wine = datasets.load_wine()
    X = wine.data
    y = wine.target
    class_names = wine.target_names

    # Training set test set .
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

    # SVM linear kernel   .
    classifier = svm.SVC(kernel='linear', C=0.01)
    y_pred = classifier.fit(X_train, y_train).predict(X_test)
    
    np.set_printoptions(precision=2)

    # confusion matrix  .
    ## 1.  plot_confusion_matrix    None .
    plot_confusion_matrix(y_test, y_pred, classes=class_names)

    #   confusion matrix  .
    ## 2.  plot_confusion_matrix    None .
    plot_confusion_matrix(y_test, y_pred, classes=class_names, normalize=True)


# Confusion matrix 
def plot_confusion_matrix(y_true, y_pred, classes, normalize=False, cmap=plt.cm.OrRd):
                          
    title = ""
    if normalize:
        title = 'Normalized confusion matrix'
    else:
        title = 'Confusion matrix'

    ## 3. Confusion matrix      .
    cm = confusion_matrix(y_true, y_pred)

    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        #        1     .
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    print(title, ":\n", cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # label 45   
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # confusion matrix   
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    plt
#     plt.savefig('confusion matrix.png')

if __name__ == "__main__":
    main()
```

#### sklearn.metrics  
- roc_curve(y_true, y_pred) -> fpr, tpr :    y    y      , FPR(False Positive Rate), TPR(True Positive Rate) .

- auc(fpr, tpr) : fpr, tpr   ROC Curve     .

-----------------------------
y     . (hint : .shape function)

roc_curve()  y_test i , y_score i     fpr[i], tpr[i]  .

auc()  fpr[i], tpr[i]      ROC & AUC  .

  y          ROC curve .   (area) ,        .

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp

def main():
    #   .
    wine = datasets.load_wine()
    # X Y  .
    X = wine.data
    y = wine.target
    # class  .
    y = label_binarize(y, classes=[0, 1, 2])
    
    
    # y shape(data dimension) .
    print(y.shape)
    
    ## 1. y     .
    n_classes = y.shape[1]
    print(n_classes)
    
    #  noise       .
    random_state = np.random.RandomState(0)
    n_samples, n_features = X.shape
    X = np.c_[X, random_state.randn(n_samples, 200 * n_features)]
    # Training / test data .
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
    
    # SVM      .
    random_state = np.random.RandomState(0)
    classifier = OneVsRestClassifier(svm.SVC(kernel='linear', probability=True, random_state=random_state))
    
    y_score = classifier.fit(X_train, y_train).decision_function(X_test)
    #   ROC Curve      .
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # ROC curve     .
    for i in range(n_classes):
        ## 2. roc_curve()  y_test i , y_score i     fpr[i], tpr[i]  .
        fpr[i], tpr[i], _ = roc_curve(y_test[:,i], y_score[:,i])
        
        ## 3. auc()  fpr[i], tpr[i]      ROC & AUC  .
        roc_auc[i] = auc(fpr[i], tpr[i])
    print (roc_auc)
    for nth_class in range(n_classes):
        plot_ROC_curve(fpr, tpr, roc_auc, nth_class)
    
# ROC curve    .
def plot_ROC_curve(fpr, tpr, roc_auc, nth_class):

    plt.figure()
    lw = 2
    
    color_name = ''
    if nth_class == 0:
        color_name = 'red'
    elif nth_class == 1 :
        color_name = 'darkorange'
    else :
        color_name = 'yellow'
        
    plt.plot(fpr[nth_class], tpr[nth_class], color=color_name,
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[nth_class])
    plt.plot([0, 1], [0, 1], color='green', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Class %s ROC(Receiver Operating Characteristic) Curve' %nth_class)
    plt.legend(loc="lower right")
#     plt.savefig('roc curve.png')
    
if __name__ == "__main__":
    main()
```

```python
import numpy as np
from sklearn.metrics import f1_score
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split

def main():
	
    ## y_true, y_pred   .
    y_true = ["apple", "banana", "cherry", "grape", "banana", "banana", "apple", "grape"]
    y_pred = ["grape", "banana", "cherry", "grape", "cherry", "grape", "apple", "grape"]
    
    # 1. average    f1 score  .
    f1_score_macro = f1_score(y_true, y_pred, average='macro')
    f1_score_micro = f1_score(y_true, y_pred, average='micro')
    f1_score_weighted = f1_score(y_true, y_pred, average='weighted')
    f1_score_None = f1_score(y_true, y_pred, average=None)
    
    print("F1_Score_macro = ", f1_score_macro)
    print("F1_Score_micro = ", f1_score_micro)
    print("F1_Score_weighted = ", f1_score_weighted)
    print("F1_Score_None = ", f1_score_None)
    
    ## Wine data SVM   y_true, y_pred  .
    y_true2, y_pred2 = get_data_for_F1_Score()
    
    # 2. average    f1 score  .
    f1_score_macro2 = f1_score(y_true2, y_pred2, average='macro')
    f1_score_micro2 = f1_score(y_true2, y_pred2, average='micro')
    f1_score_weighted2 = f1_score(y_true2, y_pred2, average='weighted')
    f1_score_None2 = f1_score(y_true2, y_pred2, average=None)
    
    print("\nF1_Score_macro2 = ", f1_score_macro2)
    print("F1_Score_micro2 = ", f1_score_micro2)
    print("F1_Score_weighted2 = ", f1_score_weighted2)
    print("F1_Score_None2 = ", f1_score_None2)
    
    
def get_data_for_F1_Score():
    # Wine data set  X, y  .
    wine = datasets.load_wine()
    X = wine.data
    y = wine.target
    class_names = wine.target_names

    # Training set test set .
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

    # SVM linear kernel   .
    classifier = svm.SVC(kernel='linear', C=0.01)
    y_pred = classifier.fit(X_train, y_train).predict(X_test)
    
    np.set_printoptions(precision=2)
    
    return y_test, y_pred


if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets, svm, metrics
from random import randint


def main():
	# digit data set .
    digits = datasets.load_digits()

    # digit data set  .
    n_samples = len(digits.images)
    data = digits.images.reshape((n_samples, -1))

    # SVM Classifier .
    classifier = svm.SVC(gamma=0.001)

    # 1. n_samples   digits.data   .
    classifier.fit(digits.data[:n_samples//2], digits.target[:n_samples//2])

    # 2.   test data set    y .
    expected = digits.target[n_samples//2:]
    predicted = classifier.predict(digits.data[n_samples//2:])
    
    #   digit data     confusion matrix .
    print("Confusion matrix results :\n\t- row : real(expected) 0 ~ 9 label\n\t- column : predicted 0 ~ 9 label\n\n%s\n" 
            % metrics.confusion_matrix(expected, predicted))
           
    #print ((metrics.confusion_matrix(expected, predicted).T[3])[3]/sum(metrics.confusion_matrix(expected, predicted).T[3]))
    #print( metrics.recall_score(expected, predicted, average=None)[2])
    # 3. Confusion matrix 3 index  precision !
    index_3_precision = metrics.precision_score(expected, predicted, average=None)[3]
    
    # 4. Confusion matrix 3 index  recall !
    index_3_recall = metrics.recall_score(expected, predicted, average=None)[3]
    
    #   4         .
    draw_digit_images(digits, n_samples, predicted)
    
    
    return index_3_precision, index_3_recall
    


##    .
def draw_digit_images(digits, n_samples, predicted):
    number = randint(0, n_samples // 2 - 4)

    images_and_labels = list(zip(digits.images[n_samples // 2:], 
        digits.target[n_samples // 2:]))
    
    for index, (image, label) in enumerate(images_and_labels[number: number + 4]):
        plt.subplot(2, 4, index + 1)
        plt.axis('off')
        plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
        plt.title('Training: %i' % label)

    images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted))
    for index, (image, prediction) in enumerate(images_and_predictions[number: number + 4]):
        plt.subplot(2, 4, index + 5)
        plt.axis('off')
        plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
        plt.title('Prediction: %i' % prediction)

#     plt.savefig('digit.png')
    plt
    
if __name__ == "__main__":
    main()
```
